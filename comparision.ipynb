```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/final_result.csv")

# Prepare data for seaborn
df_f1 = df[["Model", "ROUGE1_F1", "ROUGE2_F1", "ROUGEL_F1"]]
df_f1 = df_f1.melt(id_vars="Model", var_name="Metric", value_name="Score")

plt.figure(figsize=(12,6))
sns.barplot(data=df_f1, x="Model", y="Score", hue="Metric")

plt.title("ROUGE F1 Score Comparison Across Models")
plt.xlabel("Models")
plt.ylabel("F1 Score")
plt.yscale("log")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

precision

```python
df_prec = df[["Model", "ROUGE1_Precision", "ROUGE2_Precision", "ROUGEL_Precision"]]
df_prec = df_prec.melt(id_vars="Model", var_name="Metric", value_name="Score")

plt.figure(figsize=(12,6))
sns.barplot(data=df_prec, x="Model", y="Score", hue="Metric")

plt.title("ROUGE Precision Comparison Across Models")
plt.xlabel("Models")
plt.ylabel("Precision Score")
plt.yscale("log")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
